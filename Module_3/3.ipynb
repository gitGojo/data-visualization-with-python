```python
# Module 3
# Checking for missing values

import pandas as pd
import numpy as np

df = pd.DataFrame({"Age": [12,14,np.nan,42],
                   "Marks": [12,np.nan,45,np.nan]}
                   )
print(df)
print()
# isnull()
print(df.isnull()) # return True in case of Null Values
print()
# Print count of columns which are null
print(df.isnull().sum())
print()
# Print the count of rows which are empty
print(df.isnull().sum(axis=1))
print()
# print the count of empty cells
print(df.isnull().sum().sum())
print()
# Find the percentage of empty cells in a column
print(df.isnull().sum()/len(df)*100)
```

```python
# Handling missing values
# 1) dropna - drop rows with missing values

import pandas as pd
import numpy as np

df = pd.DataFrame({"Name":["Alice","Bob",None,"Delta"],
                   "Age" : [13,np.nan,12,14]})
df = df.dropna()
print(df)
```

```python
# 2) fillna - fills the missing values with a value
import pandas as pd
import numpy as np

df = pd.DataFrame({"Name":["Alice","Bob",None,"Delta"],
                   "Age" : [13,np.nan,12,14]})
print("Filling missing values with 0")
df = df.fillna(0)
print(df)
print()
```

```python
print("Filling missing values with mean")
df["Age"]= df["Age"].fillna(df["Age"].mean())
print(df)
```

```python
import pandas as pd
import numpy as np

s = pd.Series([10,20,np.nan,45,60]) # (20+45)/2 = 32.5 
s = s.fillna(s.median())
print(s)
```

```python
import pandas as pd
import numpy as np

s = pd.Series([10,20,20,np.nan])
s = s.fillna(s.mode()[0])
print(s)
```

```python
# 4) Interpolation - guesses the missing values based on the values around it
import pandas as pd
import numpy as np

s = pd.Series([1,np.nan,3,4,5])
s = s.interpolate()
print(s)
```

```python
# Cleaning up csv file using converters
import pandas as pd
import numpy as np

def clean_price(val):
    try:
        return float(str(val).replace("$","").replace(",","").strip())
    except ValueError:
        return None

def clean_name(name):
    return str(name).strip().title()

df = pd.read_csv("Sales1.csv",
                 converters={"Price": clean_price,"Name": clean_name})

print(df)
```

```python
# Cleaning up using students data

import pandas as pd
import numpy as np

df = pd.DataFrame({"Name":["Alice","NA","Bob","Charlie"],
                   "Age":[14,-1,23,45],
                   "Marks":[23,45,-1,89]})
df.to_csv("Students_details.csv",index=False)

df_read = pd.read_csv("Students_details.csv",na_values=[-1,"NA"])
print(df_read)
print()
df_read["Age"] = df_read["Age"].fillna("-")
df_read["Marks"] = df_read["Marks"].fillna(df_read["Marks"].median())
df_read["Name"] = df_read["Name"].fillna("Unknown")
print(df_read)
print()
print(f"Average marks : {df_read["Marks"].mean()}")
print(f"Top scorer is {df_read.loc[df_read["Marks"].idxmax(),"Name"]}")
print()
print(df_read.describe())

```

```python
# Cleanup in csv
import pandas as pd

df = pd.read_csv("Sales2.csv",na_values=["NA",-1])
# Inspect the values
df.head()
# Clean up column name
df.columns = df.columns.str.title()
print(df)
print()
# Clean up values
df["Name"]=df["Name"].str.title() # whenver you are using title() use str before that
df["Price"]=pd.to_numeric(df["Price"],errors = "coerce")
df["Date"]=pd.to_datetime(df["Date"],errors="coerce")
print(df)
# drop invalid rows
df = df.dropna()
print()
# reset index
df.reset_index(drop=True,inplace=True)
print(df)
```

```python
'''Types of joins
 1) Outer join - Join all columns in A and B
 2) Inner join - Join matching columns in A and B
 3) Left join - Join all columns in A and matching columns in B 
 use pd.merge() for all '''

import pandas as pd

employees = pd.DataFrame({"Name":["Alice","Bob","Charlie","Delta"],
                  "DeptID":[10,20,30,10],})
department = pd.DataFrame({"DeptID":[10,20],
                  "DeptName":["HR","Finance"]})

# Outer join
outer_join = pd.merge(employees,department,on="DeptID",how="outer")
print(outer_join)
print()
#inner join
inner_join = pd.merge(employees,department,on="DeptID",how="inner")
print(inner_join)
print()
# Left join
left_join = pd.merge(employees,department,on="DeptID",how="left")
print(left_join)
print()
# Right join
right_join = pd.merge(employees,department,on="DeptID",how="right")
print(right_join)
```
