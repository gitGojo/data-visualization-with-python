```python
#Series
## Creating a Series with Index From Ndarray
import pandas as pd
import numpy as np
s = pd.Series(np.random.randn(4),index=['a','b','c','d'])
print(s)
```

```python
## Slicing and Operations on Series

### Slicing Data
import pandas as pd
import numpy as np
s= pd.Series(np.random.randn(5))
print(s)
print("SLicing")
print(s[:3])
x = s[0]
print("The first element is %.4f "%x) 

'''
Why does Python use % twice?
Because one % describes the placeholder, and the other % performs the replacement.

"The first element is %.4f" % x
                         ↑      ↑
                  Placeholder   Replace with x'''

print("Multiple indexes can also be printed")
print(s[[4,2,0]]) # REMEMBER TO USE TWO [[]] 
```

```python
#SAMPLE OPERATIONS WITH SERIES
import pandas as pd
import numpy as np
s = pd.Series(np.random.randn(4))
print(s)
print("The numbers greater than the Average %.4f is" % s.mean())
print(s[s > s.mean()])
print("The numbers greater than the median %.2f is " %s.median())
print(s[s>s.median()])
print("The exponent value of each number is ")
ex = np.exp(s)
print(ex)
```

```python
# Vectorized Operations and Label Alignment with Series
import pandas as pd
import numpy as np
s = pd.Series([1,2,3,4],index=['a','b','c','d'])
print(s)
print(s+s)
print(s*s)
x = np.exp(s)
print(x)
print(s[1:]+s[:-1]) 
'''s[1: ] means [s[1] -> b,s[2] -> c,s[3]->d ] 
  s [:-1] means all excluding the last element so it includes label a,b,c 
  so this operation s[1:]+s[:-1] adds only the matching labels and rest all unmatched labels are not added and therefore NaN'''
```

```python
# Assigning New Columns in Method Chains
# Another way to add column is by assign()
# Syntax is df.assign(column_name = value)
import pandas as pd
data = {"one":[1,2,3],"two": [2,3,4]}
df = pd.DataFrame(data)
print(df)
print()
# Adding another column using assign()
df = df.assign(three = lambda x : x["one"]+x["two"])
print(df)
print()
# Adding another column called four
df = df.assign(four = lambda x: x["three"]+3)
print(df)
print()
# Column can also be modified 
df = df.assign(three = lambda x: x["one"])
print(df)
```

```python
## Indexing and Selecting Data Frames

### Data Frame Indexing Methods
# Select column - results in a series
print(df["one"])
print()
# Select row by integer location - DataFrame
print(df.iloc[2])
print()
# Slice rows
print(df["two"][:-1])
print()
# Select rows by Boolean vector
print(df[df["one"]>1])
```

```python
# Regular expressions
# look for @uct.ac.za in the file
txt = open("Emails.txt")
for line in txt:
    line = line.rstrip()
    if not '@uct.ac.za' in line:
        continue
    else:
        print(line)
```

```python
#Extract Lines Starting with Pattern From
txt = open("Emails.txt")
for line in txt:
    line = line.strip()
    if line.startswith("From"):
        print(line)
```

```python
# Extract email without regular expression that is extract only stephen.marquard@uct.ac.za without From:
txt = open("Emails.txt")
for line in txt:
    line = line.rstrip()
    if line.startswith("From: "):
        line = line.split()
        print(line[1])
```

```python
# Do the same operation with regex , find pattern which starts with X and ends with :
import re
txt = open("Emails.txt")
for line in txt:
    line=line.rstrip()
    y= re.findall('X.+:',line)
    print(y)

```

```python
# remove space characters
import re
txt = open("Emails.txt")
for line in txt:
    line = line.rstrip()
    y = re.findall(r'X-\S+:',line)
    print(y)
```

```python
# Extracting Numerical Values 
import re
txt = open("Emails.txt")
for line in txt:
    line = line.rstrip()
    y = re.findall(r'[0-9]+',line)
    print(y)
```

```python
# Extracting Specific Characters
import re
txt = open("Emails.txt")
for line in txt:
    line = line.rstrip()
    y = re.findall('[Fgp]',line)
    print(y)
```

```python
# Greedy technique - Give the maximum possible 
import re
x = "From : blah blah : yes"
y = re.findall('^F.+:',x) # read till the end 
print(y)
```

```python
# Non greedy technique - give as minimum as possible
import re
x = "From : blah blah : yes"
y = re.findall('^F.+?:',x) # read till the first :
print(y)
```

```python
# Extract E-mail Using Regular Expressions:
import re
txt = "From ossama.embarak@hct.ac.ae Sat Jun 5 08:14:16 2018"
extract_text = re.findall(r'\S+@\S+',txt)
print(extract_text)
```

```python
# Extract mail and username without regular expressions
txt = "From ossama.embarak@hct.ac.ae Sat Jun 5 08:14:16 2018"
pos_at = txt.find('@')
at_pos = txt.find(' ',pos_at)
mail = txt[pos_at+1:at_pos]
print(mail)
user_pos = txt.find(' ')
user_name = txt[user_pos+1:pos_at]
print(user_name)
print(f"Email is {user_name+'@'+mail}")
```

```python
# Extract Specific Parts (Listing 4-14 from Textbook)
import re
txt = "From ossama.embarak@hct.ac.ae Sat Jun 5 08:14:16 2018"
y = re.findall('@([^ ]+)',txt)
print(y)
```

```python
# Escape Characters Example:
# $ is a regular expression suppose if you want to use it as regular symbol , you can \$ 
mystr = 'We just received $10.00 for cookies and $20.23 for juice'
y = re.findall(r'\$[0-9]+',mystr)
print(y)
```
